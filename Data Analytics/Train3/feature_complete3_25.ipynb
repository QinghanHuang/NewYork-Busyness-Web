```python
df = df.drop(['congestion_surcharge'], axis=1)
```

```python
df_2019_2018 = panda.concat([df_2018, df], ignore_index=True)
```

```python
df_2019_2018 =   df_2019_2018.sort_values(by='dropoff_datetime_hour')
```

```python
df_2019_2018['dropoff_datetime_hour'] = panda.to_datetime(df_2019_2018['dropoff_datetime_hour'])
```

```python
df_2019_2018['year'] =  df_2019_2018['dropoff_datetime_hour'].dt.year
```

```python
df_2019_2018['busyness'] = df_2019_2018['number_taxis_DOLocationID_for_this_hour'] / df_2019_2018['DOLocation_area']
```

```python
df_2019_2018 =   df_2019_2018.drop('number_taxis_DOLocationID_for_this_hour',  axis=1)
```

```python
df_2019_2018 = df_2019_2018.drop(['dropoff_datetime_hour'],  axis=1)
```

```python
import numpy as nump


df_2019_2018['busyness'] = df_2019_2018['busyness'].astype(float)


bins = nump.percentile(df_2019_2018['busyness'], [0, 20, 40, 60, 80, 100])


labels = [1, 2, 3, 4, 5]


df_2019_2018['busyness'] = panda.cut(df_2019_2018['busyness'], bins=bins, labels=labels, include_lowest=True)

```

```python

yx =   ['DOLocationID',   'VendorID',    'RatecodeID',   'payment_type',
                      'dropoff_hour',   'dropoff_day_of_week',    'dropoff_day_of_month',
                                'dropoff_month']


df_2019_2018 = panda.get_dummies(df_2019_2018, columns=yx)


print(df.head())
```
