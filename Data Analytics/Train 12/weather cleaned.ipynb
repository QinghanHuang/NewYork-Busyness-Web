### Handling Missing Values

```python
for columns in weather.columns:
    print(f"{columns}: {(weather[columns].isna().mean())* 100}")
```

Observations
--------------------------------------
As observed above several columns have more than 70% of their contents missing. Also we do not think this column would have any impact on our proposed model bases on domain knowledge, hence I would be dropping these features

```python
#Dropping Features where missing values are over 50%
for column in weather:
    if (weather[column].isna().mean() * 100) > 50:
        weather.drop([column], axis = 1, inplace = True)
```

```python
for columns in weather.columns:
    print(f"{columns}: {(weather[columns].isna().mean())* 100}")
```

```python
weather[weather["visibility"].isna()]
```

```python
weather.fillna(weather['visibility'].median(), inplace = True)
```

```python
for columns in weather.columns:
    print(f"{columns}: {(weather[columns].isna().mean())* 100}")
```

#### Observations
----------------------------------

The visibility feature has less than 1% of its contents missing. Since this feature is skewed, I would be replacing missing values with the median. 

Another observation, there seem to be duplicate values in the dt (i.e. datetime feature), and all rows with nan visibility seems to have this issue. However, this could be because the weather changed twice during the hour, and was therefore recorded twice. more investigation needed

### Checking For constant features

```python
for column in weather:
    print(f"{column}, has {weather[column].nunique()} values")
```

```python
#dropping constant features
for column in weather:
    if weather[column].nunique() < 2:
        weather.drop([column], axis = 1, inplace = True)
```

In this section we dropped constant features.
- both the latitude and longitude are constant features and do not provide any substantial information, except the location, which we are already aware of
