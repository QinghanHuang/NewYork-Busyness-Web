```python
import pandas as panda
import numpy as nump


df['busyness'] = df['busyness'].astype(float)



bins = nump.percentile(df['busyness'], [0, 20, 40, 60, 80, 100])


labels = [1, 2, 3, 4, 5]


df['busyness'] = panda.cut(df['busyness'], bins=bins, labels=labels, include_lowest=True)

```

```python

yx =   ['DOLocationID',   'VendorID',    'RatecodeID',   'payment_type',
                      'dropoff_hour',   'dropoff_day_of_week',    'dropoff_day_of_month',
                                'dropoff_month', 'weather_description', 'weather_main']


df = panda.get_dummies(df, columns=yx)


print(df.head())
```
